## 3.1. å‡å€¼å‘é‡çš„æ£€éªŒ

### 3.1.1. å•ç»„æ ·æœ¬å‡å€¼å‘é‡å‡å®šçš„æ£€éªŒ
å•ç»„æ ·æœ¬å‡å€¼å‘é‡æ£€éªŒä¸å•ç»„æ ·æœ¬çš„å‡å€¼æ£€éªŒç±»ä¼¼ï¼Œåªä¸è¿‡å‡å®šçš„å¯¹è±¡ä»ä¸€ä¸ªæ•°å€¼å˜æˆäº†ä¸€ä¸ªå‘é‡ã€‚å¯¹äºè¿™ä¸ªé—®é¢˜ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨å•æ ·æœ¬hotelling T2æ£€éªŒã€‚

è®¾$\boldsymbol{x}_1,\cdots ,\boldsymbol{x}_n\,$æ¥ä¸€ä¸ª$p$å…ƒæ­£æ€åˆ†å¸ƒçš„ä¸€ä¸ªæ ·æœ¬ã€‚å¯¹äºå•æ ·æœ¬hotelling T2æ£€éªŒï¼Œæˆ‘ä»¬åšå¦‚ä¸‹ä¸¤ä¸ªå‡è®¾ï¼š
$$
H_0:\boldsymbol{\mu }=\boldsymbol{\mu }_0\leftrightarrow H_1:\boldsymbol{\mu }\ne \boldsymbol{\mu }_0\,
$$
æ³¨æ„ï¼Œè¿™é‡Œçš„$\boldsymbol{\mu }$æ˜¯åŠ ç²—çš„ï¼Œä»£è¡¨å‘é‡çš„æ„æ€ã€‚

æ£€éªŒç»Ÿè®¡é‡ä¸ºï¼š
$$
Test\,\,statistics=\frac{n-p}{p\left( n-1 \right)}T^2\,\,, T^2=n\left( \boldsymbol{\bar{x}}-\boldsymbol{\mu }_0 \right) '\boldsymbol{S}^{-1}\left( \boldsymbol{\bar{x}}-\boldsymbol{\mu }_0 \right) 
$$
å…¶ä¸­ï¼Œ$\boldsymbol{S}$æ ·æœ¬çš„æ— ååæ–¹å·®çŸ©é˜µ

æ£€éªŒç»Ÿè®¡é‡æœä»åˆ†å¸ƒ
$$
Test\,\,statistics\sim F\left( p,n-p \right) 
$$
på€¼è®¡ç®—å…¬å¼ä¸º
$$
pvalue=P\left( F>Test\,\,statistics \right) 
$$
æ³¨æ„ï¼Œå°½ç®¡å¤‡æ‹©å‡è®¾çš„ç¬¦å·æ˜¯$\ne$ï¼Œä½†è¿™é‡Œpå€¼å¹¶ä¸æ˜¯åŒè¾¹på€¼ï¼Œå¤šå…ƒæ£€éªŒä¸­på€¼å½¢å¼çš„åˆ¤æ–­å¹¶ä¸èƒ½æ²¿ç”¨ä¸€å…ƒå‡å€¼æ£€éªŒçš„åˆ¤æ–­æ–¹æ³•ã€‚

**Example.5** æˆ‘ä»¬æƒ³ç ”ç©¶æŸåœ°åŒºå†œæ‘2å²ç”·å©´çš„å‘è‚²çŠ¶å†µæ˜¯å¦è¾¾åˆ°å›½å®¶å‚è€ƒæ ‡å‡†ï¼Œäºæ˜¯æŠ½å–äº†6åç”·å©´å¹¶æµ‹é‡äº†ä»–ä»¬çš„èº«é«˜$x_1$ã€èƒ¸å›´$x_2$ã€ä¸ŠåŠè‡€å›´$x_3$ã€‚å·²çŸ¥è¿™ä¸‰ä¸ªæŒ‡æ ‡çš„å›½å®¶å‚è€ƒæ ‡å‡†å‡å€¼ä¸º$\boldsymbol{\mu }_0=\left( 90,58,16 \right) '$ï¼Œ6åç”·å©´çš„æ•°æ®å¦‚ä¸‹æ‰€ç¤ºï¼Œé—®èƒ½å¦è®¤ä¸ºè¯¥åœ°åŒºå†œæ‘ç”·å©´çš„å‘è‚²çŠ¶å†µè¾¾åˆ°å›½å®¶æ ‡å‡†ï¼Ÿ

ç”±äºscipyåŒ…ä¸­å¹¶æ²¡æœ‰Hotelling T2çš„apiï¼Œå› æ­¤æˆ‘ä»¬éœ€è¦ä½¿ç”¨ä¸Šè¿°çš„æ£€éªŒç»Ÿè®¡é‡åŠå…¶åˆ†å¸ƒè‡ªå·±ç¼–å†™å‡è®¾æ£€éªŒã€‚

```python
from scipy.stats import f
```

```python
# æ„é€ ä¸€ä¸ªå‡½æ•°ï¼Œå¸®åŠ©æˆ‘ä»¬è¿›è¡Œå•æ ·æœ¬Hotelling T2æ£€éªŒ
def multi_checkmean(data:pd.DataFrame,mean:np.ndarray,confidence=0.05):
    '''
    data:å¾…æ¯”è¾ƒçš„çŸ©é˜µï¼Œæ ¼å¼æœ€å¥½ä¸ºpd.DataFrame()ï¼Œä»¥åˆ—ä¸ºå˜é‡ï¼Œä»¥è¡Œä¸ºæ ·æœ¬ã€‚
    mean:å‡å®šçš„å‡å€¼å‘é‡ï¼Œæ ¼å¼ä¸ºnumpy.ndarrayã€‚
    '''

    # è®¡ç®—æ£€éªŒç»Ÿè®¡é‡
    S=np.cov(data.T)
    S_inv=np.linalg.inv(S)
    x_bar=np.mean(data).values.T
    T2=len(data)*np.dot((x_bar-mean).T,np.dot(S_inv,x_bar-mean))
    Test_statistics=T2*(len(data)-len(mean))/(len(mean)*(len(data)-1))

    # è®¡ç®—på€¼
    pvalue=f.sf(Test_statistics,len(mean),len(data)-len(mean))

    # è¾“å‡ºæ ·æœ¬çš„å‡å€¼å‘é‡
    print('æ ·æœ¬å‡å€¼å‘é‡ä¸ºï¼š{}'.format(x_bar))
    print('ç›®æ ‡å‡å€¼å‘é‡ä¸ºï¼š{}'.format(mean))

    # æ¯”è¾ƒpå€¼ä¸æ˜¾è‘—æ€§æ°´å¹³
    if pvalue<confidence:
        print('åœ¨æ˜¾è‘—æ€§æ°´å¹³{0:}ä¸‹ï¼Œæ ·æœ¬å‡å€¼å‘é‡ä¸ç­‰äºå‡å®šçš„å‡å€¼å‘é‡ã€‚(p={1:.4f})'.format(confidence,pvalue))
    else:
        print('åœ¨æ˜¾è‘—æ€§æ°´å¹³{0:}ä¸‹ï¼Œæ ·æœ¬å‡å€¼å‘é‡ç­‰äºå‡å®šçš„å‡å€¼å‘é‡ã€‚(p={1:.4f})'.format(confidence,pvalue))
    return pvalue
```

### 3.1.2. ä¸¤ç»„æ ·æœ¬ä¹‹é—´çš„å‡å€¼å‘é‡ç›¸ç­‰æ€§æ£€éªŒ

ä¸å•å˜é‡çš„åŒæ ·æœ¬å‡å€¼æ£€éªŒä¸€æ ·ï¼ŒåŒæ ·æœ¬å‡å€¼å‘é‡çš„æ£€éªŒåŒæ ·åˆ†ä¸ºç»„åˆ«é—´ç‹¬ç«‹ä¸ç»„åˆ«æˆå¯¹ä¸¤ç§æƒ…å†µï¼Œä¸”åˆ¤åˆ«ä¸¤è€…çš„æ–¹æ³•ä¹Ÿæ˜¯ç›¸åŒçš„ï¼Œæ­¤å¤„ä¸å†èµ˜è¿°ã€‚æˆ‘ä»¬ç›´æ¥å¼€å§‹æ„å»ºå‡è®¾æ£€éªŒã€‚

#### ç»„åˆ«é—´ç‹¬ç«‹
**Example.6** æˆ‘ä»¬æƒ³çŸ¥é“3å‘¨å²çš„ç”·å¥³å©´å„¿çš„èº«ä½“æŒ‡æ ‡æ˜¯å¦å­˜åœ¨æ˜¾è‘—å·®å¼‚ï¼Œäºæ˜¯å„æŠ½å–äº†ç”·å¥³å„15åå©´å„¿ï¼Œå¹¶æµ‹é‡äº†ä»–ä»¬çš„èº«é«˜$X_1$ä¸ä½“é‡$X_2$ï¼Œé—®ï¼šåœ¨æ˜¾è‘—æ€§æ°´å¹³0.05ä¸‹èƒ½å¦è®¤ä¸ºä¸¤è€…å­˜åœ¨æ˜¾è‘—å·®å¼‚ï¼Ÿ

ç”±äºä¸¤ä¸ªæ ·æœ¬æ€§åˆ«ä¸åŒï¼Œå› æ­¤æˆ‘ä»¬å¯ä»¥è®¤ä¸ºä¸¤ä¸ªæ ·æœ¬ç›¸äº’ç‹¬ç«‹ã€‚

è®¾ä¸¤ä¸ªæ ·æœ¬$\boldsymbol{x}_1,\cdots ,\boldsymbol{x}_{n1}\,$å’Œ$\boldsymbol{y}_1,\cdots ,\boldsymbol{y}_{n2}\,$åˆ†åˆ«æ¥è‡ªä¸¤ä¸ª$p$å…ƒæ­£æ€åˆ†å¸ƒï¼Œæˆ‘ä»¬åšå¦‚ä¸‹ä¸¤ä¸ªå‡è®¾ï¼š
$$
H_0:\boldsymbol{\mu_1 }=\boldsymbol{\mu }_2\leftrightarrow H_1:\boldsymbol{\mu_1 }\ne \boldsymbol{\mu }_2\,
$$
æ£€éªŒç»Ÿè®¡é‡ä¸ºï¼š
$$
Test\,\,statistics=\frac{n_1+n_2-p-1}{p\left( n_1+n_2-2 \right)}T^2\,\,, T^2=\frac{n_1n_2}{n_1+n_2}(\overline{\boldsymbol{x}}-\overline{\boldsymbol{y}})^{'}\boldsymbol{S}_{p}^{-1}(\overline{\boldsymbol{x}}-\overline{\boldsymbol{y}}) , \boldsymbol{S}_p=\frac{\left( n_1-1 \right) \boldsymbol{S}_1+\left( n_2-1 \right) \boldsymbol{S}_2}{n_1+n_2-2}
$$
æ£€éªŒç»Ÿè®¡é‡æœä»åˆ†å¸ƒï¼š
$$
Test\,\,statistics\sim F\left( p,n_1+n_2-p-1 \right) 
$$
på€¼å½¢å¼ä¸ºï¼š
$$
pvalue=P\left( F>Test\,\,statistics \right) 
$$

```python
data=pd.read_excel('./data/ex6.xlsx')
# å°†ä¸¤ä¸ªç‹¬ç«‹æ ·æœ¬åˆ†å¼€æˆä¸¤ç»„
group_boy=data[data['gender']=='b'].drop('gender',axis=1) # å»é™¤æ€§åˆ«æ ‡ç­¾åˆ—ï¼Œåªä¿ç•™æ•°å€¼åˆ—
group_girl=data[data['gender']=='g'].drop('gender',axis=1)
group_girl
```

```python
def multi_unparied_data(group1:pd.DataFrame,group2:pd.DataFrame,confidence=0.05):
    # è®¡ç®—æ£€éªŒç»Ÿè®¡é‡
    n1=len(group1)
    n2=len(group2)
    p=np.shape(group1)[1] # å˜é‡ç»´åº¦
    mean1=np.mean(group1).values.T
    mean2=np.mean(group2).values.T
    S1=np.cov(group1.T)
    S2=np.cov(group2.T)
    Sp=((n1-1)*S1+(n2-1)*S2)/(n1+n2-2)
    T2=n1*n2*(mean1-mean2).T@np.linalg.inv(Sp)@(mean1-mean2)/(n1+n2)
    Test_statistics=(n1+n2-p-1)*T2/(p*(n1+n2-2))

    # è®¡ç®—på€¼
    pvalue=f.sf(Test_statistics,p,n1+n2-p-1)

    # æ¯”è¾ƒpå€¼ä¸æ˜¾è‘—æ€§æ°´å¹³
    if pvalue<confidence:
        print('åœ¨æ˜¾è‘—æ€§æ°´å¹³{0:}ä¸‹ï¼Œä¸¤ç»„æ ·æœ¬æ‰€åœ¨æ€»ä½“çš„å‡å€¼å‘é‡ä¸ç›¸ç­‰ã€‚(p={1:.4f})'.format(confidence,pvalue))
    else:
        print('åœ¨æ˜¾è‘—æ€§æ°´å¹³{0:}ä¸‹ï¼Œä¸¤ç»„æ ·æœ¬æ‰€åœ¨æ€»ä½“çš„å‡å€¼å‘é‡ç›¸ç­‰ã€‚(p={1:.4f})'.format(confidence,pvalue))
    return pvalue
```

```python
multi_unparied_data(group_boy,group_girl)
```

#### æˆå¯¹ç»„åˆ«

å¤šå…ƒå‡å€¼å‘é‡æ£€éªŒä¸­ï¼Œæˆå¯¹æ£€éªŒçš„åŸç†ä¸ä¸€å…ƒå‡å€¼æ£€éªŒçš„æˆå¯¹æ£€éªŒåŸç†æ˜¯ç›¸ä¼¼çš„ï¼Œæœ¬è´¨ä¸Šåšä¸¤ä¸ªå‡å€¼å‘é‡ä¹‹å·®ä¸é›¶å‘é‡çš„ä¹‹é—´çš„å•æ ·æœ¬å‡å€¼å‘é‡æ£€éªŒã€‚

è®¾$\left( \boldsymbol{x}_i,\boldsymbol{y}_i \right) \,\,, i=1,\cdots ,n $ä¸ºæœä»$p$å…ƒæ­£æ€åˆ†å¸ƒçš„æˆå¯¹æ•°æ®ï¼Œä»¤ä»–ä»¬ç›¸å‡å¾—ï¼š
$$
\boldsymbol{d}_i=\boldsymbol{x}_i-\boldsymbol{y}_i
$$
æ£€éªŒç»Ÿè®¡é‡ä¸ºï¼š
$$
Test\,\,statistics=\frac{n-p}{p\left( n-1 \right)}T^2\,\,, T^2=n\boldsymbol{\bar{d}}'\boldsymbol{S}_{\boldsymbol{d}}^{-1}\boldsymbol{\bar{d}}
$$
æ£€éªŒç»Ÿè®¡é‡æœä»åˆ†å¸ƒï¼š
$$
Test\,\,statistics\sim F\left( p,n-p \right) 
$$
på€¼è®¡ç®—å…¬å¼ä¸º
$$
pvalue=P\left( F>Test\,\,statistics \right) 
$$

```python
def multi_paried_data(group1:pd.DataFrame,group2:pd.DataFrame,confidence=0.05):
    # è®¡ç®—æ£€éªŒç»Ÿè®¡é‡å€¼
    n=len(group1)
    p=np.shape(group1)[1]
    mean1=np.mean(group1).values.T
    mean2=np.mean(group2).values.T
    d=group1-group2
    d_mean=mean1-mean2
    Sd=np.cov(d.T)
    Sd_inv=np.linalg.inv(Sd)
    T2=n*d_mean.T@Sd_inv@d_mean
    Test_statistics=(n-p)*T2/(p*(n-1))
    
    # è®¡ç®—på€¼
    from scipy.stats import f

    pvalue=f.sf(Test_statistics,p,n-p)

    # æ¯”è¾ƒpå€¼ä¸æ˜¾è‘—æ€§æ°´å¹³
    if pvalue<confidence:
        print('åœ¨æ˜¾è‘—æ€§æ°´å¹³{0:}ä¸‹ï¼Œä¸¤ç»„æ ·æœ¬æ‰€åœ¨æ€»ä½“çš„å‡å€¼å‘é‡ä¸ç›¸ç­‰ã€‚(p={1:.4f})'.format(confidence,pvalue))
    else:
        print('åœ¨æ˜¾è‘—æ€§æ°´å¹³{0:}ä¸‹ï¼Œä¸¤ç»„æ ·æœ¬æ‰€åœ¨æ€»ä½“çš„å‡å€¼å‘é‡ç›¸ç­‰ã€‚(p={1:.4f})'.format(confidence,pvalue))
    return pvalue

    
```
